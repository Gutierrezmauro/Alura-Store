```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head(10)
```

## **1-Analisis de facturacion**

```python
ventas = tienda['Precio'].sum()
print('La suma del valor total de los productos es:', round(ventas))

ventas2 = tienda2['Precio'].sum()
print('La suma del valor total de los productos es:', round(ventas2))

ventas3 = tienda3['Precio'].sum()
print('La suma del valor total de los productos es:', round(ventas3))

ventas4 = tienda4['Precio'].sum()
print('La suma del valor total de los productos es:', round(ventas4))
```

## **2-Ventas por categoria**

```python
#tienda.groupby('Categoría del Producto')[['Producto']].value_counts()
```

```python
cat1 = tienda['Categoría del Producto'].value_counts().head(3)
cat2 = tienda2['Categoría del Producto'].value_counts().head(3)
cat3 = tienda3['Categoría del Producto'].value_counts().head(3)
cat4 = tienda4['Categoría del Producto'].value_counts().head(3)

print('Las 3 categorias con mas ventas son:')
print(cat1)
print('Las 3 categorias con mas ventas son:')
print(cat2)
print('Las 3 categorias con mas ventas son:')
print(cat3)
print('Las 3 categorias con mas ventas son:')
print(cat4)
```

## **3-Clasificacion promedio de la tienda**

```python
tienda.head()
```

```python
promedio1 = round(tienda['Calificación'].mean(),2)
print(f'El promedio de calificación de la tienda 1 es: {promedio1}')
```

```python
promedio2 = round(tienda2['Calificación'].mean(),2)
print(f'El promedio de calificación de la tienda 2 es: {promedio2}')
```

```python
promedio3 = round(tienda3['Calificación'].mean(),2)
print(f'El promedio de calificación de la tienda 3  es: {promedio3}')
```

```python
promedio4 = round(tienda4['Calificación'].mean(),1)
print(f'El promedio de calificación de la tienda 4 es: {promedio4}')
```

## **4-Productos mas y menos vendidos**

```python
prod_max = tienda['Producto'].value_counts().head(2)
prod_min = tienda['Producto'].value_counts().tail(2)

prod_max2 = tienda2['Producto'].value_counts().head(2)
prod_min2 = tienda2['Producto'].value_counts().tail(2)

prod_max3 = tienda3['Producto'].value_counts().head(2)
prod_min3 = tienda3['Producto'].value_counts().tail(2)

prod_max4 = tienda4['Producto'].value_counts().head(2)
prod_min4 = tienda4['Producto'].value_counts().tail(2)

productos_totales = pd.concat([prod_max,prod_min,prod_max2,prod_min2,prod_max3,prod_min3,prod_max4,prod_min4])

productos_totales.plot(kind='bar',figsize=(12,8), color='pink', xlabel= 'Productos', ylabel= 'Ventas')
# Tipos de graficos ('line', 'bar', 'barh', 'kde', 'density', 'area', 'hist', 'box', 'pie', 'scatter', 'hexbin')
```

## **5-Envio promedio de la tienda**

```python
tienda.head(10)
```

```python
cost_envio = round(tienda['Costo de envío'].mean())
print('El promedio del costo de los envios de la tienda 1 es:', cost_envio)

cost_envio2 = round(tienda2['Costo de envío'].mean())
print('El promedio del costo de los envios de la tienda 2 es:', cost_envio2)

cost_envio3 = round(tienda3['Costo de envío'].mean())
print('El promedio del costo de los envios de la tienda 3 es:', cost_envio3)

cost_envio4 = round(tienda4['Costo de envío'].mean())
print('El promedio del costo de los envios de la tienda 4 es:', cost_envio4)
```
